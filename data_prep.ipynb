#  ASHTANGRUDYAM CONLLU -> JSON

```python
import pandas as pd
import json
```

```python
with open("/home/shivraj-pg/DEPNECT/ashtangrudyam.conllu", "r", encoding="utf-8") as f:
    text = f.read()
```

```python
asth_data = text.split("\n\n")
```

```python
type(asth_data), len(asth_data)
```

```python
print(asth_data[0])
```

```python
json_asthang = []

for sentence in asth_data:
    try:
        temp = {"sentence": "", "tokens":[], "compounds":[]}
        sent = sentence.split("\n")
        # print(sent)
        temp["sentence"] = sent[2].replace("# text = ", "").replace("- ", "")
        temp["tokens"].append(sent[2].replace("# text = ", "").replace("-", "").split(" "))
        
        temp["compounds"] = "\n".join(sent[3:])
        # print("#######################################")    
        json_asthang.append([temp["sentence"], {"tokens":temp['tokens'][0], "compounds":temp["compounds"]}])
    
    except Exception as e:
        with open("error.txt", "a", encoding="utf-8") as ef:
            ef.write(sentence)
            ef.write(f"\n{e}")
            ef.write("##################################################\n\n")
            continue
    
```

```python
df = pd.DataFrame(json_asthang, columns=["sentence", "gold"])
```

```python
df.head()
```

```python
df['gold'][0]
```

```python
print(df['gold'][0]['compounds'])
```

```python
df.to_csv("/home/shivraj-pg/DEPNECT/conllu-style-csv/asthangrudyam.csv", index=False)
```

Conllu converter

```python
df_list = []
idx = 0
for sentence in sentences:
    try:
        temp = {"tokens":[], "compounds":""}
        words = sentence.split("\n")
        temp['compounds'] = "\n".join(words)
        for word in words:
            subword = word.split('\t')
            # print("subword[1]", subword[1])
            temp["tokens"].append(subword[1])
            
        df_list.append([" ".join(temp['tokens']),temp])   
        # print(words)
        # print("######################################\n")

    except Exception as e:
        with open("/home/shivraj-pg/DEPNECT/error.txt", "a", encoding="utf-8") as ef:
            ef.write(sentence)
            ef.write(f"ERROR: {e}")
            ef.write("##################################################\n\n")
            continue
```

```python
df = pd.DataFrame(df_list, columns=["sentence", "gold"])
```
