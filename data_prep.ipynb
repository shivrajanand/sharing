#  ASHTANGRUDYAM CONLLU -> JSON

```python
import pandas as pd
import json
```

```python
with open("/home/shivraj-pg/DEPNECT/ashtangrudyam.conllu", "r", encoding="utf-8") as f:
    text = f.read()
```

```python
asth_data = text.split("\n\n")
```

```python
asth_data[0]
```

```python
json_asthang = []

for sentence in asth_data:
    try:
        temp = {"sentence": "", "tokens":[], "compounds":[]}
        sent = sentence.split("\n")
        # print(sent)
        temp["sentence"] = sent[2].replace("# text = ", "").replace("- ", "")
        temp["tokens"].append(sent[2].replace("# text = ", "").replace("-", "").split(" "))
        
        for words in sent[3:]:
            temp_comp_data = {"span":[], "label":""}
            word = words.split('\t')
            # print(word)
            temp_comp_data["span"].append(word[0])
            temp_comp_data["span"].append(word[3])
            temp_comp_data["label"] = word[-1]
            
            temp["compounds"].append(temp_comp_data)
        # print("#######################################")    
        json_asthang.append(temp)
    
    except Exception as e:
        with open("error.txt", "a", encoding="utf-8") as ef:
            ef.write(sentence)
            ef.write(f"\n{e}")
            ef.write("##################################################\n\n")
            continue
    
```

```python
len(json_asthang)
```

```python
with open("ashtangrudyam.json", "w", encoding="utf-8") as f:
    json.dump(json_asthang, f, ensure_ascii=False, indent=4)
```

```python
with open("ashtangrudyam.json", "r", encoding="utf-8") as file:
    data = json.load(file)
```

# Without-Context-Coarse -> Json

```python
import pandas as pd
import json
```

```python
with open("/home/shivraj-pg/DEPNECT/Without_Context_Coarse/test.conllu", "r", encoding="utf-8") as f:
    text = f.read()
```

```python
sentences = text.split('\n\n')
```

```python
print(sentences[0])
```

```python
dict_list = []

for sentence in sentences:
    try:
        temp = {"tokens":[], "compounds":[]}
        words = sentence.split("\n")
        
        for word in words:
            subword = word.split('\t')
            # print("subword[1]", subword[1])
            temp["tokens"].append(subword[1])
            temp["compounds"].append(
                {
                    "span": [subword[0], subword[6]],
                    "label": subword[7]
                }
            )
        dict_list.append(temp)   
        # print(words)
        # print("######################################\n")

    except Exception as e:
        with open("/home/shivraj-pg/DEPNECT/error.txt", "a", encoding="utf-8") as ef:
            ef.write(sentence)
            ef.write(f"ERROR: {e}")
            ef.write("##################################################\n\n")
            continue
```

```python
with open("Without_context_coarse_test.json", "w", encoding="utf-8") as f:
    json.dump(dict_list, f, ensure_ascii=False, indent=4)
```
